```python
# Importing Dependencies
import pandas as pd
import numpy as np
```

```python
# Bringing in Uncleaned CSV's
csv1 = "Resources/201809-citibike-tripdata_1.csv"
csv2 = "Resources/201909-citibike-tripdata_1.csv"
bike_trip_2018 = pd.read_csv(csv1)
bike_trip_2019 = pd.read_csv(csv2)
```

```python
# Viewing the 2018 Data
bike_2018_df = bike_trip_2018
bike_2018_df.head(2)
```

```python
# Viewing the 2019 Data
bike_2019_df = bike_trip_2019
bike_2019_df.head(2)
```

```python
# Deleting Columns
del bike_2018_df['usertype']
del bike_2019_df['usertype']
```

```python
bike_2018_df.head(5)
bike_2019_df.head(5)
```

```python
# Checking Type of Values
bike_2018_df.dtypes
bike_2019_df.dtypes
```

```python
# Changing The Type of Values for the Gender Column
bike_2018_df['gender'] = bike_2018_df['gender'].astype(str)
bike_2019_df['gender'] = bike_2019_df['gender'].astype(str)
```

```python
# Changing the Numeric Gender Values to Actual Gender Values For The 2018 CSV
gender_check_2018 = bike_2018_df['gender']
changed_2018_male = gender_check_2018.replace('1', 'Male')
changed_2018_female = changed_2018_male.replace('2', 'Female')
changed_2018_unknown = changed_2018_female.replace('0', 'Unknown')
bike_2018_df['gender'] = changed_2018_unknown
bike_2018_df
```

```python
# Changing the Numeric Gender Values to Actual Gender Values For The 2019 CSV
gender_check_2019 = bike_2019_df['gender']
changed_2019_male = gender_check_2019.replace('1', 'Male')
changed_2019_female = changed_2019_male.replace('2', 'Female')
changed_2019_unknown = changed_2019_female.replace('0', 'Unknown')
bike_2019_df['gender'] = changed_2019_unknown
bike_2019_df
```

```python
# Calculating Biker's Age's 
current_year = 2019

birth_year_2018 = bike_2018_df['birth year']
age = current_year - birth_year_2018
bike_2018_df['age'] = age

birth_year_2019 = bike_2019_df['birth year']
age = current_year - birth_year_2019
bike_2019_df['age'] = age
```

```python
# Rearranging the Columns
bike_2018_df_organized = bike_2018_df[["bikeid",
                                       "gender",
                                       "birth year",
                                       "age",
                                       "tripduration",
                                       "start station id",
                                       "start station name",
                                       "start station latitude",
                                       "start station longitude",
                                       "starttime",
                                       "end station id",
                                       "end station name", 
                                       "end station latitude",
                                       "end station longitude",
                                       "stoptime"]]

bike_2019_df_organized = bike_2019_df[["bikeid",
                                       "gender",
                                       "birth year",
                                       "age",
                                       "tripduration",
                                       "start station id",
                                       "start station name",
                                       "start station latitude",
                                       "start station longitude",
                                       "starttime",
                                       "end station id",
                                       "end station name", 
                                       "end station latitude",
                                       "end station longitude",
                                       "stoptime"]]

bike_2018_df_organized.head(5)
bike_2019_df_organized.head(5)
```

```python
# Renaming the Columns
bike_2018_renamed_df = bike_2018_df_organized.rename(columns={"bikeid":"Bike ID", 
                                                              "gender":"Gender",
                                                              "birth year":"Birth Year",
                                                              "age":"Age",
                                                              "tripduration":"Trip Duration",
                                                              "start station id":"Start Station ID",
                                                              "start station name":"Start Station Name",
                                                              "start station latitude":"Start Station Latitude",
                                                              "start station longitude":"Start Station Longitude",
                                                              "starttime":"Start Time",
                                                              "end station id":"End Station ID",
                                                              "end station name":"End Station Name",
                                                              "end station latitude":"End Station Latitude",
                                                              "end station longitude":"End Station Longitude",
                                                              "stoptime":"Stop Time"})

bike_2019_renamed_df = bike_2019_df_organized.rename(columns={"bikeid":"Bike ID", 
                                                              "gender":"Gender",
                                                              "birth year":"Birth Year",
                                                              "age":"Age",
                                                              "tripduration":"Trip Duration",
                                                              "start station id":"Start Station ID",
                                                              "start station name":"Start Station Name",
                                                              "start station latitude":"Start Station Latitude",
                                                              "start station longitude":"Start Station Longitude",
                                                              "starttime":"Start Time",
                                                              "end station id":"End Station ID",
                                                              "end station name":"End Station Name",
                                                              "end station latitude":"End Station Latitude",
                                                              "end station longitude":"End Station Longitude",
                                                              "stoptime":"Stop Time"})



bike_2018_renamed_df
bike_2019_renamed_df
```

```python
# Keep only the first 50,000 rows of each DataFrame
bike_2018_50k = bike_2018_renamed_df.head(50000)
bike_2019_50k = bike_2019_renamed_df.head(50000)

# Outputting both files to CSV in Outputs folder
bike_2018_50k.to_csv("Outputs/09_2018_citibike_tripdata_cleaned.csv", index=False, header=True)
bike_2019_50k.to_csv("Outputs/09_2019_citibike_tripdata_cleaned.csv", index=False, header=True)
```
